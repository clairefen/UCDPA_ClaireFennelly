```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Happiness = pd.read_csv('world-happiness-report-2021.csv')
print(Happiness.head())
```

```python
print(Happiness.isna().sum())
```

```python
Happiness.rename(columns = {'Country name': 'country'}, inplace=True)
```

```python
# no NAs in the Happiness df, however if needed this is how I would drop them
drop_columns = Happiness.dropna()
print(drop_columns.shape)
```

```python
Happiness.head()
```

```python
# Using iloc to see the top 20 countries and their freedom to make life choices to compare to bottom 20
MostHappy = Happiness.iloc[0:20, [0,2,8,9]]
MostHappy.head()
```

```python
# Using iloc to see the bottom 20 countries and their freedom to make life choices 
MostUnhappy = Happiness.iloc[129:150, [0, 2, 8,9]]
MostUnhappy.head()
```

```python
# Column mean scores
Happiness.mean()
```

```python
# finding the mean of each value in order to map the scores by region
Happiness_Region = Happiness.groupby('Regional indicator').mean()
Happiness_Region.head()
```

```python
Alc_rates = pd.read_csv('gapminder_alcohol.csv')
print(Alc_rates.head())
```

```python
# To demonstrate groupby and sum, seeing what countries have the highest alcohol consumption
Alc_rates_grouped = Alc_rates.groupby(['country']).sum()['alcconsumption'].sort_values(ascending=False)
```

```python
print(Alc_rates_grouped.head())
```

```python
drop_columns = Alc_rates.dropna()
print(drop_columns.shape)
```

```python
# Merging the Happiness df with alcohol/suicide rates dataframe
happiness_alc_merge = pd.merge(Happiness, Alc_rates, on='country', how='inner')
```

```python
happiness_alc_merge.head()
```

```python
Mosthappy_alc = happiness_alc_merge.iloc[0:30, [0, 2, 20, 22]]
Mosthappy_alc.head()
```

```python
Mostunhappy_alc = happiness_alc_merge.iloc[128:149, [0, 2, 20, 22]]
Mostunhappy_alc.head()
```

```python
happiness_alc_merge.mean()
```

```python
# Demonstrating the use of numpy array

alc_consump_arr = np.array(happiness_alc_merge['alcconsumption'])
print(alc_consump_arr)
```

```python
# how many countries are consuming alcohol above the average?

alc_consump_arr > 7.072985
```

```python
# Creating a function to calculate the difference between upper and lower whiskers to assess variability
def whiskerdif(num1, num2):
    answer = num1 - num2
    return answer

whiskerdif(7.904, 7.780)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data= Happiness_Region, x='Ladder score', y='Regional indicator', marker='o')
ax.set_title('Average happiness per world region')
ax.set(xlabel='Happiness rating', ylabel='World Region')
plt.savefig("image.png")
```

```python
fig, ax = plt.subplots()
sns.barplot(data= MostHappy, x='country', y='Ladder score')
ax.set_title('Happiest countries in the world')
plt.xticks(rotation=90)
ax.set(xlabel='Country', ylabel='Happiness rating')
plt.savefig("image10.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(data= MostUnhappy, x='country', y='Ladder score', color='grey')
ax.set_title('Most unhappy countries in the world')
sns.color_palette("magma", as_cmap=True)
plt.xticks(rotation=90)
ax.set(xlabel='Country', ylabel='Happiness rating')
plt.savefig("image11.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize= (10, 3))
x = Mostunhappy_alc['country']
y1 = Mostunhappy_alc['Ladder score']
y2 = Mostunhappy_alc['suicideper100th']
plt.xticks(rotation=90)
ax.set(xlabel='Country', ylabel='happiness rating/suicide rate')
ax.set_title('Suicide rates vs happiness score for lowest countries (that we have data for)')
ax.plot(x, y1)
ax.plot(x, y2)
plt.savefig("image12.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize= (20, 3))
x = Mosthappy_alc['country']
y1 = Mosthappy_alc['Ladder score']
y2 = Mosthappy_alc['suicideper100th']
ax.set_title('Happiness vs Suicide rate in top 30 countries')
plt.xticks(rotation=90)
ax.plot(x, y1)
ax.plot(x, y2)
plt.savefig("image13.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
